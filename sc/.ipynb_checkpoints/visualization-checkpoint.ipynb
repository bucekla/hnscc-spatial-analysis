```python
# --- Gene Name Matching Logic (Revised with Suffix Check) ---
print("\n--- Checking Marker Lists Against adata.var_names (with suffix check) ---")

adata_genes = adata.var_names.tolist()
adata_genes_upper_set = {gene.upper() for gene in adata_genes}
adata_gene_map_upper_to_original = {gene.upper(): gene for gene in adata_genes}

# Dictionaries to store results per list
found_markers_by_list = {name: [] for name in marker_lists}
missing_markers_by_list = {name: [] for name in marker_lists}
mapped_via_suffix = {} # Keep track of mappings: {original_marker: mapped_adata_gene}
all_found_markers_set = set() # Keep track of all unique found markers across lists

for list_name, marker_list in marker_lists.items():
    print(f"\nChecking list: '{list_name}' ({len(marker_list)} markers)")
    if not marker_list:
        print("  List is empty, skipping.")
        continue

    current_list_found = set() # Use set for intermediate findings within a list

    for marker in marker_list:
        if not isinstance(marker, str) or not marker: continue

        marker_upper = marker.upper()
        found = False
        original_adata_gene = None

        # 1. Try direct match (case-insensitive)
        if marker_upper in adata_genes_upper_set:
            original_adata_gene = adata_gene_map_upper_to_original[marker_upper]
            current_list_found.add(original_adata_gene)
            all_found_markers_set.add(original_adata_gene)
            found = True
            # print(f"  Direct match: '{marker}' -> '{original_adata_gene}'") # Optional debug print

        # 2. If direct match failed, try removing suffix 'a' or 'b'
        if not found and (marker.endswith('a') or marker.endswith('b')):
            base_name = marker[:-1]
            base_name_upper = base_name.upper()
            # print(f"  -> Trying base name: '{base_name}' (from '{marker}')") # Optional debug print
            if base_name_upper in adata_genes_upper_set:
                original_adata_gene = adata_gene_map_upper_to_original[base_name_upper]
                current_list_found.add(original_adata_gene)
                all_found_markers_set.add(original_adata_gene)
                found = True
                mapped_via_suffix[marker] = original_adata_gene # Record the mapping
                # print(f"    Suffix match: '{marker}' -> '{original_adata_gene}'") # Optional debug print


        # 3. If still not found, add to missing list for this category
        if not found:
            missing_markers_by_list[list_name].append(marker)
            # Heuristic check remains useful for other non-matching patterns
            # if marker.islower() or '.' in marker or marker.endswith(('.L', '.S')):
            #      mismatch_markers_by_list[list_name].append(marker) # This dict isn't used in reporting now


    # Store the unique found genes for this list
    found_markers_by_list[list_name] = sorted(list(current_list_found))
    print(f"  Found {len(found_markers_by_list[list_name])} unique corresponding markers in adata for this list.")


# --- Report Summary ---
print("\n--- Summary of Marker Checking (with suffix mapping) ---")
total_found_unique = len(all_found_markers_set)
print(f"Total unique markers found across all lists: {total_found_unique}")

if mapped_via_suffix:
     print("\nMarkers mapped via suffix removal (Original -> Found in adata):")
     for orig, mapped in sorted(mapped_via_suffix.items()):
          print(f"  '{orig}' -> '{mapped}'")

for list_name in marker_lists:
    print(f"\nList: '{list_name}'")
    found_list = found_markers_by_list[list_name]
    missing_list = sorted(list(set(missing_markers_by_list[list_name]))) # Unique missing

    if found_list:
        print(f"  Found ({len(found_list)}): {found_list}")
    else:
        print("  Found (0)")

    if missing_list:
        # Report only those missing *after* trying suffix removal
        final_missing = [m for m in missing_list if m not in mapped_via_suffix]
        # Also check base names weren't found (more robust check needed if base names differed across suffixes)
        print(f"  Missing ({len(final_missing)} after suffix check): {sorted(final_missing)}")
    else:
        print("  Missing (0)")

# --- Prepare Consolidated Lists for Plotting ---
all_rics_found = sorted(list(set(
    found_markers_by_list.get("RICs_CSV", []) +
    found_markers_by_list.get("RICs_manual", []) +
    found_markers_by_list.get("RICs_diff", [])
)))
all_rocs_found = sorted(list(set(
    found_markers_by_list.get("ROCs_manual", []) +
    found_markers_by_list.get("ROCs_diff", [])
)))
all_combined_found = sorted(list(all_found_markers_set)) # Already unique

print("\n--- Consolidated Marker Lists for Plotting ---")
print(f"All RICs found ({len(all_rics_found)}): {all_rics_found}")
print(f"All ROCs found ({len(all_rocs_found)}): {all_rocs_found}")
print(f"All Combined found ({len(all_combined_found)}): {all_combined_found}")
```

```python
print("\n--- Comparing Each Leiden Resolution with 'cell.type' Metadata ---")
resolutions_to_test = [0.5, 1.0, 1.5,2.5]
# 2. Loop through resolutions again to compare and plot
for res in resolutions_to_test:
    cluster_key = f'leiden_integrated_res{res}'
    print(f"\n--- Comparison for Resolution: {res} (key='{cluster_key}') ---")

    if 'cell.type' not in adata.obs.columns:
        print("Column 'cell.type' not found in adata.obs. Skipping comparison for this resolution.")
        continue # Skip to the next resolution

    if cluster_key not in adata.obs.columns:
        print(f"Cluster key '{cluster_key}' not found in adata.obs. Skipping comparison for this resolution.")
        continue # Skip to the next resolution

    # --- Create Contingency Table ---
    ct = pd.crosstab(adata.obs[cluster_key], adata.obs['cell.type'])
    # print(f"Crosstab: {cluster_key} (rows) vs cell.type (columns)")
    # print(ct)

    # --- Visualize as Heatmap (Counts) ---
    plt.figure(figsize=(10, max(5, len(ct)*0.4))) # Adjust height based on number of clusters
    sns.heatmap(ct, annot=True, fmt='d', cmap='viridis', linewidths=.5)
    plt.title(f'Leiden Cluster (res={res}) vs Metadata cell.type Counts')
    plt.ylabel(f'Leiden Cluster (res={res})')
    plt.xlabel('Metadata cell.type')
    plt.show()

    # --- Calculate and Visualize Proportions (% Leiden Cluster in cell.type) ---
    # Handle cases where a cluster might have zero cells (though unlikely with Leiden)
    ct_sum = ct.sum(axis=1)
    # Avoid division by zero if a cluster somehow ends up empty
    ct_norm_by_leiden = ct.apply(lambda r: r / r.sum() * 100 if r.sum() > 0 else r, axis=1)
    # print(f"\nNormalized Crosstab (% of {cluster_key} Cells in each cell.type):")
    # print(ct_norm_by_leiden.round(1)) # Print rounded percentages

    plt.figure(figsize=(10, max(5, len(ct)*0.4))) # Adjust height based on number of clusters
    sns.heatmap(ct_norm_by_leiden, annot=True, fmt='.1f', cmap='viridis', linewidths=.5)
    plt.title(f'% of Leiden Cluster (res={res}) Cells in each Metadata cell.type')
    plt.ylabel(f'Leiden Cluster (res={res})')
    plt.xlabel('Metadata cell.type')
    plt.show()


print("\n--- Comparison across resolutions complete ---")
print("Examine the heatmaps to see how cluster purity/composition changes with resolution.")
```

```python
leiden_resolution = 2.5 # Or your chosen resolution
leiden_key = f'leiden_integrated_res{leiden_resolution}'
output_plot_dir = "./dotplots/" # Define directory for saving plots
os.makedirs(output_plot_dir, exist_ok=True) # Create if needed

# Check if the specific Leiden key exists
if leiden_key not in adata.obs.columns:
    print(f"ERROR: Leiden clustering key '{leiden_key}' not found in adata.obs.")
else:
    print(f"\n--- Generating Dotplots vs Leiden Clusters (res={leiden_resolution}) ---")
    print(f"Using Leiden clustering key: '{leiden_key}'")

    # --- Prepare Markers and Categories for Combined Plot ---
    print("\nPreparing markers for combined plot with separators...")

    # Define categories and the markers found within them
    # Important: Order matters for plotting and lines
    categories_in_order = ["RICs_CSV", "RICs_manual", "RICs_diff", "ROCs_manual", "ROCs_diff"]
    markers_by_cat = {cat: found_markers_by_list.get(cat, []) for cat in categories_in_order}

    # Create the final ordered list of unique markers for the combined plot
    combined_markers_ordered = []
    category_boundaries = {} # Store y-index where each category ends
    current_index = -1
    for cat in categories_in_order:
        unique_markers_in_cat = sorted([m for m in markers_by_cat[cat] if m not in combined_markers_ordered])
        if unique_markers_in_cat:
            combined_markers_ordered.extend(unique_markers_in_cat)
            current_index += len(unique_markers_in_cat)
            category_boundaries[cat] = current_index # Mark end index

    print(f"Total unique markers in combined ordered list: {len(combined_markers_ordered)}")
    # print(f"Category end indices: {category_boundaries}") # Debug print

    # Remove the last boundary as we only draw lines *between* categories
    categories_to_draw_lines_after = list(category_boundaries.keys())[:-1]


    # --- Plot 3: All Combined Markers with Separator Lines ---
    if combined_markers_ordered:
        print("\nPlotting Dotplot for All Combined Markers with Separators...")
        plot_filename_combined = f"dotplot_combined_separated_vs_leiden{leiden_resolution}.png"
        plot_save_path = os.path.join(output_plot_dir, plot_filename_combined)

        # Generate the dotplot object - DO NOT SHOW, get the axes grid
        # Use standard_scale='var' maybe? Or leave default.
        dp = sc.pl.dotplot(adata, combined_markers_ordered, groupby=leiden_key,
                         title=f'All Markers by Category vs Leiden Clusters (res={leiden_resolution})',
                         return_fig=True, # Get the figure object
                         show=False) # VERY IMPORTANT: Do not show yet

        # Access the main axes object where the dots are plotted
        # Dotplot returns a DotPlot object; access axes grid then main axes
        # This structure might change slightly between scanpy versions
        try:
             # Common way to access the axes grid
             axes_grid = dp.get_axes()
             # The main heatmap/dotplot axes is usually the first one
             ax = axes_grid['mainplot_ax']

             print("Adding separator lines...")
             # Draw horizontal lines BELOW each category (except the last)
             for cat in categories_to_draw_lines_after:
                  boundary_index = category_boundaries[cat]
                  # axhline y-coordinate is typically between indices
                  line_y = boundary_index + 0.5
                  ax.axhline(line_y, color='grey', linestyle='--', linewidth=1)
                  print(f"  Line drawn below category '{cat}' at y={line_y:.1f}")

             # --- Save and Show the Modified Figure ---
             print(f"Saving plot to: {plot_save_path}")
             # Use the figure object returned by dotplot
             dp.savefig(plot_save_path, dpi=150, bbox_inches='tight')
             # Show the modified plot
             dp.show()
             # Close the figure object after saving/showing
             # plt.close(dp.get_figure()) # Close if needed, dp.show might handle it

        except Exception as e:
             print(f"Error accessing axes or drawing lines: {e}")
             print("Showing unmodified dotplot instead.")
             # Fallback to showing the unmodified plot
             dp_fallback = sc.pl.dotplot(adata, combined_markers_ordered, groupby=leiden_key,
                                         title=f'All Markers vs Leiden Clusters (res={leiden_resolution})',
                                         show=False)
             dp_fallback.show()
             # plt.close(dp_fallback.get_figure())

    else:
        print("No Combined markers found to plot.")


    # --- Plot 1 & 2: Separate RICs and ROCs Dotplots (No separators needed here) ---
    # (Code remains the same as your previous correct version for these)

    # Plot 1: All RICs Markers
    all_rics_found = sorted(list(set(found_markers_by_list.get("RICs_CSV", []) + found_markers_by_list.get("RICs_manual", []) + found_markers_by_list.get("RICs_diff", []))))
    if all_rics_found:
        print("\nPlotting Dotplot for All RICs Markers...")
        plot_filename_rics = f"dotplot_rics_vs_leiden{leiden_resolution}.png"
        sc.pl.dotplot(adata, all_rics_found, groupby=leiden_key,
                     title=f'All Found RICs Markers vs Leiden Clusters (res={leiden_resolution})',
                     show=False,
                     save=f"_{plot_filename_rics}"
                     )
        plt.clf(); sc.pl.dotplot(adata, all_rics_found, groupby=leiden_key, title=f'All Found RICs Markers vs Leiden Clusters (res={leiden_resolution})'); plt.show()
    else:
        print("No RICs markers found to plot.")

    # Plot 2: All ROCs Markers
    all_rocs_found = sorted(list(set(found_markers_by_list.get("ROCs_manual", []) + found_markers_by_list.get("ROCs_diff", []))))
    if all_rocs_found:
        print("\nPlotting Dotplot for All ROCs Markers...")
        plot_filename_rocs = f"dotplot_rocs_vs_leiden{leiden_resolution}.png"
        sc.pl.dotplot(adata, all_rocs_found, groupby=leiden_key,
                     title=f'All Found ROCs Markers vs Leiden Clusters (res={leiden_resolution})',
                     show=False,
                     save=f"_{plot_filename_rocs}"
                     )
        plt.clf(); sc.pl.dotplot(adata, all_rocs_found, groupby=leiden_key, title=f'All Found ROCs Markers vs Leiden Clusters (res={leiden_resolution})'); plt.show()
    else:
        print("No ROCs markers found to plot.")


print("\n--- Dotplot generation complete ---")
```
