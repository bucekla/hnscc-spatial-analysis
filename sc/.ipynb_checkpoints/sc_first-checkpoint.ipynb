```python
import scanpy as sc
import os
import shutil
import re
from collections import defaultdict
from tempfile import TemporaryDirectory
import tempfile

# Your data directory
# data_dir = "/Users/ladislavbucek/Desktop/UK/Mgr./labgenexp/spatial_project/data/sc/GSE139324_RAW"
scratch_dir = os.getenv("SCRATCHDIR", "/tmp")  # fallback to /tmp if not set
tmp_dir = tempfile.mkdtemp(dir=scratch_dir)
# Group file paths by sample
file_groups = defaultdict(dict)
for fname in os.listdir(data_dir):
    if fname.endswith(('.tsv', '.mtx')):
        match = re.match(r"(GSM\d+_[^_]+_[^_]+_[^_]+)", fname)
        if match:
            sample_id = match.group(1)
            fpath = os.path.join(data_dir, fname)
            if "barcodes" in fname:
                file_groups[sample_id]['barcodes'] = fpath
            elif "genes" in fname:
                file_groups[sample_id]['genes'] = fpath
            elif "matrix" in fname:
                file_groups[sample_id]['matrix'] = fpath

# Load each sample using temporary directories
adatas = []
with TemporaryDirectory() as tmp_root:
    for sample_id, paths in file_groups.items():
        if all(k in paths for k in ['barcodes', 'genes', 'matrix']):
            print(f"Loading {sample_id}")

            # Create temp folder and copy files in
            scratch_dir = os.getenv("SCRATCHDIR", "/tmp")  # fallback to /tmp if not set
            tmp_dir = tempfile.mkdtemp(dir=scratch_dir)
            os.makedirs(tmp_dir, exist_ok=True)

            # Copy files with expected Cell Ranger v2 names
            shutil.copy(paths['genes'], os.path.join(tmp_dir, 'genes.tsv'))
            shutil.copy(paths['barcodes'], os.path.join(tmp_dir, 'barcodes.tsv'))
            shutil.copy(paths['matrix'], os.path.join(tmp_dir, 'matrix.mtx'))

            # Load with Scanpy
            ad = sc.read_10x_mtx(tmp_dir, var_names='gene_symbols', make_unique=True)
            ad.obs['sample'] = sample_id
            adatas.append(ad)
        else:
            print(f"Skipping {sample_id}: missing one or more files")

# # Merge all samples
# adata = adatas[0].concatenate(
#     adatas[1:], 
#     batch_key="sample", 
#     batch_categories=[a.obs['sample'][0] for a in adatas]
# )

# # Save for reuse
# adata.write("merged_all_samples.h5ad")
```

```python
normalized_adatas = []
# adata_qc = sc.concat(list(adatas_by_sample.values()), label='sample', keys=samples, index_unique='-')

for sample in adata.obs['sample'].unique():
    ad = adata[adata.obs['sample'] == sample].copy()
    
    # Normalize & log1p
    sc.pp.normalize_total(ad, target_sum=1e4)
    sc.pp.log1p(ad)

    # Store it
    normalized_adatas.append(ad)
adata_norm = sc.concat(normalized_adatas, label='sample', keys=[a.obs['sample'].iloc[0] for a in normalized_adatas], index_unique='-')

```

integration

```python
import scanpy.external as sce

sc.pp.highly_variable_genes(adata_norm, batch_key='sample')
adata_hvg = adata_norm[:, adata_norm.var.highly_variable].copy()
sc.pp.scale(adata_hvg)
sc.tl.pca(adata_hvg)
sce.pp.harmony_integrate(adata_hvg, key='sample')
sc.pp.neighbors(adata_hvg, use_rep='X_pca_harmony')
sc.tl.umap(adata_hvg)
for res in [0.2, 0.5, 1.0, 1.5]:
    sc.tl.leiden(adata_hvg, resolution=res, key_added=f'leiden_{res}',flavor='igraph',directed=False,n_iterations=2)

```

```python

sc.pl.umap(adata_hvg, color='sample', title='Samples after Harmony Integration')
# sc.pl.umap(adata_hvg, color='leiden', title='Leiden Clusters after Harmony')
for res in [0.2, 0.5, 1.0, 1.5]:
    n_clusters = adata_hvg.obs[f'leiden_{res}'].nunique()
    print(f"Resolution {res}: {n_clusters} clusters")

sc.pl.umap(adata_hvg, color=['leiden_0.2', 'leiden_0.5', 'leiden_1.0', 'leiden_1.5'], wspace=0.4)

```

annotation

```python
#genes for ACT annotation
import pandas as pd

# Run marker detection
sc.tl.rank_genes_groups(adata_hvg, groupby='leiden', method='t-test', n_genes=40)

# Extract top genes per cluster
n_top = 40
act_lines = []

for cluster in adata_hvg.obs['leiden'].cat.categories:
    top_genes = adata_hvg.uns['rank_genes_groups']['names'][cluster][:n_top]
    genes = ",".join(top_genes)
    act_lines.append(f"cluster_{cluster}:{genes}")

# Save to file
with open("act_input_genes.txt", "w") as f:
    f.write("\n".join(act_lines))

print("✅ ACT input saved as 'act_input_genes.txt'")
```
