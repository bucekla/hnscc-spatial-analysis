```python
import scanpy as sc
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.sparse import csr_matrix
from scipy.sparse import csr_matrix
import seaborn as sns # For heatmap visualization of crosstab

```

```python
metadata_file = '/storage/praha1/home/bucekl/labgenexp/spatial_project/sc/GSE181919/GSE181919_Barcode_metadata.txt'
counts_file = '/storage/praha1/home/bucekl/labgenexp/spatial_project/sc/GSE181919/GSE181919_UMI_counts.txt'
```

```python
try:
    adata = sc.read_text(counts_file, delimiter='\t', first_column_names=True).T
    # We transpose (.T) because Scanpy prefers cells x genes (observations x variables)
    print(f"Successfully loaded counts. AnnData object shape: {adata.shape}")
    print(f"(Cells: {adata.n_obs}, Genes: {adata.n_vars})")
    print("First 5 cell names (adata.obs_names):", adata.obs_names[:5].tolist())
    print("First 5 gene names (adata.var_names):", adata.var_names[:5].tolist())
    if any('.' in name for name in adata.obs_names[:100]): # Check a sample
         print("\nAttempting to fix barcode mismatch: Replacing '.' with '-' in adata.obs_names...")
         # Use the .str accessor directly on the Index object for efficient replacement
         adata.obs_names = adata.obs_names.str.replace('.', '-', regex=False)
    else:
         print("\nBarcode names in counts matrix do not appear to contain '.' - skipping replacement.")

    # Ensure data is sparse for efficiency (read_text might load as dense)
    if not isinstance(adata.X, csr_matrix):
        adata.X = csr_matrix(adata.X)
    print(f"Count matrix type: {type(adata.X)}")

except FileNotFoundError:
    print(f"ERROR: Counts file not found at {counts_file}")
    print("Please make sure the file exists and the path is correct.")
    # Stop execution if file not found
    raise
except Exception as e:
    print(f"ERROR: Failed to load counts file with sc.read_text: {e}")
    print("Please check the format of GSE181919_UMI_counts.txt.")
    print("It should be tab-separated, with genes in rows, cells in columns.")
    print("The first column should be gene names, and the first row cell barcodes.")
    raise
```

```python
print(f"\nLoading metadata from: {metadata_file}")
try:
    metadata = pd.read_csv(metadata_file, sep='\t', index_col=0) # Assume first column is barcode index
    print(f"Successfully loaded metadata. Shape: {metadata.shape}")
    print("Metadata columns:", metadata.columns.tolist())
    print("Metadata head:\n", metadata.head())

    # --- 2. Merge Metadata with AnnData Object ---
    print("\nMerging metadata with AnnData object...")

    # Check if cell barcodes match between counts (adata.obs_names) and metadata (metadata.index)
    print(f"Cells in counts: {adata.n_obs}")
    print(f"Cells in metadata: {len(metadata)}")

    # Find common cells and filter/reorder metadata to match AnnData
    common_cells = adata.obs_names.intersection(metadata.index)
    print(f"Found {len(common_cells)} common cells between counts and metadata.")

    if len(common_cells) == 0:
        print("ERROR: No common cell barcodes found between counts and metadata!")
        print("Please check that the cell barcodes in both files match.")
        print("Example barcodes in counts:", adata.obs_names[:5].tolist())
        print("Example barcodes in metadata index:", metadata.index[:5].tolist())
        raise ValueError("Cell barcode mismatch")
    elif len(common_cells) < adata.n_obs or len(common_cells) < len(metadata):
        print("Warning: Some cells are not present in both files. Subsetting to common cells.")
        # Filter adata and metadata to keep only common cells, ordered according to adata
        adata = adata[common_cells, :].copy() # Filter AnnData (use .copy())
        metadata_filtered = metadata.loc[common_cells] # Filter and reorder metadata
        print(f"Filtered AnnData shape: {adata.shape}")
    else:
        # If all cells match, just reorder metadata to be safe
        metadata_filtered = metadata.loc[adata.obs_names]
        print("All cells match. Ordering metadata to match AnnData.")

    # Assign the filtered and ordered metadata to adata.obs
    adata.obs = metadata_filtered
    print("Successfully merged metadata into adata.obs.")
    print("adata.obs head:\n", adata.obs.head())
    # Verify
    assert all(adata.obs_names == adata.obs.index), "Mismatch between adata.obs_names and adata.obs.index after merging!"


except FileNotFoundError:
    print(f"ERROR: Metadata file not found at {metadata_file}")
    print("Please make sure the file exists and the path is correct.")
    raise
except Exception as e:
    print(f"ERROR: Failed to load or merge metadata: {e}")
    print("Please check the format of GSE181919_Barcode_metadata.txt (should be tab-separated, first column barcodes).")
    raise
```

```python
# --- Gene Name Matching Logic (Revised for Separate Lists) ---
print("\n--- Checking Marker Lists Separately Against adata.var_names ---")

adata_genes = adata.var_names.tolist()
adata_genes_upper_set = {gene.upper() for gene in adata_genes}
adata_gene_map_upper_to_original = {gene.upper(): gene for gene in adata_genes}

# Dictionaries to store results per list
found_markers_by_list = {name: [] for name in marker_lists}
missing_markers_by_list = {name: [] for name in marker_lists}
mismatch_markers_by_list = {name: [] for name in marker_lists} # Likely species mismatches
all_found_markers_set = set() # Keep track of all unique found markers across lists

for list_name, marker_list in marker_lists.items():
    print(f"\nChecking list: '{list_name}' ({len(marker_list)} markers)")
    if not marker_list:
        print("  List is empty, skipping.")
        continue

    list_found_count = 0
    for marker in marker_list:
        if not isinstance(marker, str) or not marker: # Skip NAs or empty strings
             continue
        marker_upper = marker.upper()
        if marker_upper in adata_genes_upper_set:
            original_adata_gene = adata_gene_map_upper_to_original[marker_upper]
            # Add the *actual adata gene name* to the found list
            found_markers_by_list[list_name].append(original_adata_gene)
            all_found_markers_set.add(original_adata_gene)
            list_found_count += 1
        else:
            missing_markers_by_list[list_name].append(marker)
            # Heuristic for species mismatch
            if marker.islower() or '.' in marker or marker.endswith(('a', 'b', '.L', '.S')):
                 mismatch_markers_by_list[list_name].append(marker)

    # Ensure uniqueness within each list's found markers
    found_markers_by_list[list_name] = sorted(list(set(found_markers_by_list[list_name])))
    print(f"  Found {len(found_markers_by_list[list_name])} markers in adata for this list.")

# --- Report Summary ---
print("\n--- Summary of Marker Checking ---")
total_found_unique = len(all_found_markers_set)
print(f"Total unique markers found across all lists: {total_found_unique}")
for list_name in marker_lists:
    print(f"\nList: '{list_name}'")
    found_list = found_markers_by_list[list_name]
    missing_list = missing_markers_by_list[list_name]
    mismatch_list = mismatch_markers_by_list[list_name]

    if found_list:
        print(f"  Found ({len(found_list)}): {found_list}")
    else:
        print("  Found (0)")

    if missing_list:
        print(f"  Missing ({len(missing_list)}): {sorted(missing_list)}")
        if mismatch_list:
             print(f"    Of which {len(mismatch_list)} look like non-human symbols: {sorted(mismatch_list)}")
    else:
        print("  Missing (0)")

# You now have 'found_markers_by_list' dictionary containing lists of markers
# found *in adata's format* for each of your original lists.
# Example: found_markers_by_list['RICs_diff'] will contain the diff RIC markers present in adata.
```

```python
# --- Assuming 'adata' is processed (PCA, neighbors, UMAP computed) ---
# --- Assuming 'found_markers_by_list' dictionary is populated from step 1 ---

print("\n--- Clustering, Comparison, and Annotation ---")

# 1. Perform Leiden Clustering
leiden_resolution = 1.0 # Adjust as needed
leiden_key = f'leiden_res{leiden_resolution}'
resolutions_to_test = [0.2, 0.5, 1.0, 1.5]

print(f"\n--- Running Leiden Clustering for Resolutions: {resolutions_to_test} ---")

# 1. Run Leiden for each resolution
for res in resolutions_to_test:
    cluster_key = f'leiden_{res}'
    print(f"\nRunning Leiden with resolution = {res} -> key='{cluster_key}'")
    sc.tl.leiden(
        adata,
        resolution=res,
        key_added=cluster_key,
        flavor='igraph',  # Using igraph flavor as requested
        directed=False,
        n_iterations=2
    )
    n_clusters = len(adata.obs[cluster_key].unique())
    print(f"  Found {n_clusters} clusters.")

    # Optional: Visualize this specific clustering on UMAP
    sc.pl.umap(adata, color=[cluster_key, 'cell.type'],
               title=f'Leiden (res={res}) vs Metadata cell.type',
               legend_loc='on data', legend_fontsize=8)
    plt.show()
```

```python
# 2. Loop through resolutions again to compare and plot
for res in resolutions_to_test:
    cluster_key = f'leiden_{res}'
    print(f"\n--- Comparison for Resolution: {res} (key='{cluster_key}') ---")

    if 'cell.type' not in adata.obs.columns:
        print("Column 'cell.type' not found in adata.obs. Skipping comparison for this resolution.")
        continue # Skip to the next resolution

    if cluster_key not in adata.obs.columns:
        print(f"Cluster key '{cluster_key}' not found in adata.obs. Skipping comparison for this resolution.")
        continue # Skip to the next resolution

    # --- Create Contingency Table ---
    ct = pd.crosstab(adata.obs[cluster_key], adata.obs['cell.type'])
    # print(f"Crosstab: {cluster_key} (rows) vs cell.type (columns)")
    # print(ct)

    # --- Visualize as Heatmap (Counts) ---
    plt.figure(figsize=(10, max(5, len(ct)*0.4))) # Adjust height based on number of clusters
    sns.heatmap(ct, annot=True, fmt='d', cmap='viridis', linewidths=.5)
    plt.title(f'Leiden Cluster (res={res}) vs Metadata cell.type Counts')
    plt.ylabel(f'Leiden Cluster (res={res})')
    plt.xlabel('Metadata cell.type')
    plt.show()

    # --- Calculate and Visualize Proportions (% Leiden Cluster in cell.type) ---
    # Handle cases where a cluster might have zero cells (though unlikely with Leiden)
    ct_sum = ct.sum(axis=1)
    # Avoid division by zero if a cluster somehow ends up empty
    ct_norm_by_leiden = ct.apply(lambda r: r / r.sum() * 100 if r.sum() > 0 else r, axis=1)
    # print(f"\nNormalized Crosstab (% of {cluster_key} Cells in each cell.type):")
    # print(ct_norm_by_leiden.round(1)) # Print rounded percentages

    plt.figure(figsize=(10, max(5, len(ct)*0.4))) # Adjust height based on number of clusters
    sns.heatmap(ct_norm_by_leiden, annot=True, fmt='.1f', cmap='viridis', linewidths=.5)
    plt.title(f'% of Leiden Cluster (res={res}) Cells in each Metadata cell.type')
    plt.ylabel(f'Leiden Cluster (res={res})')
    plt.xlabel('Metadata cell.type')
    plt.show()


print("\n--- Comparison across resolutions complete ---")
```

```python
# 2. Compare Leiden Clusters with Metadata Annotation ('cell.type')
print("\nComparing Leiden clusters with 'cell.type' metadata...")
if 'cell.type' in adata.obs.columns:
    # Create contingency table
    ct = pd.crosstab(adata.obs['leiden'], adata.obs['cell.type'])
    # print("Crosstab: Leiden Clusters (rows) vs Metadata cell.type (columns)")
    # print(ct)

    # Visualize as a heatmap (optional, good for many clusters/types)
    plt.figure(figsize=(10, 7))
    sns.heatmap(ct, annot=True, fmt='d', cmap='viridis', linewidths=.5)
    plt.title('Leiden Cluster vs Metadata cell.type Counts')
    plt.ylabel('Leiden Cluster')
    plt.xlabel('Metadata cell.type')
    plt.show()

    # Calculate proportions (e.g., percentage of each Leiden cluster that belongs to a metadata type)
    ct_norm_by_leiden = ct.apply(lambda r: r / r.sum() * 100, axis=1)
    # print("\nNormalized Crosstab: % of Leiden Cluster Cells in each Metadata cell.type")
    # print(ct_norm_by_leiden.round(1)) # Print rounded percentages

    plt.figure(figsize=(10, 7))
    sns.heatmap(ct_norm_by_leiden, annot=True, fmt='.1f', cmap='viridis', linewidths=.5)
    plt.title('% of Leiden Cluster Cells in each Metadata cell.type')
    plt.ylabel('Leiden Cluster')
    plt.xlabel('Metadata cell.type')
    plt.show()

else:
    print("Column 'cell.type' not found in adata.obs. Skipping comparison.")
```
